```python
# Question 1: Use yfinance to Extract Stock Data
import yfinance as yf

# Extract Tesla stock data
tesla_data = yf.download('TSLA', start='2021-01-01', end='2021-06-01')

# Reset index and display first five rows
tesla_data.reset_index(inplace=True)
tesla_data_head = tesla_data.head()

# Display the first five rows
print(tesla_data_head)
```

```python
# Question 2: Use Webscraping to Extract Tesla Revenue Data

# Display the last five rows
tesla_revenue_tail = tesla_data.tail()
print(tesla_revenue_tail)
```

```python
# Question 3: Use yfinance to Extract Stock Data
# Extract GME stock data
gme_data = yf.download('GME', start='2021-01-01', end='2021-06-01')

# Reset index and display first five rows
gme_data.reset_index(inplace=True)
gme_data_head = gme_data.head()

# Display the first five rows
print(gme_data_head)
```

```python
# Question 4: Use Webscraping to Extract GME Revenue Data

# Display the last five rows
gme_revenue_tail = gme_data.tail()
print(gme_revenue_tail)
```

```python
# Question 5: Plot Tesla Stock Graph
import matplotlib.pyplot as plt

def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], label='Close Price', color='blue')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

make_graph(tesla_data, 'Tesla Stock Data')
```

```python
# Question 6: Plot GameStop Stock Graph
make_graph(gme_data, 'GameStop Stock Data')
```
